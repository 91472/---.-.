```python
# загружаем необходимые библиотеки
import pymorphy2     # для морфологического анализа слов
import pandas as pd  # для работы с исходным файлом .csv
import re            # для работы с регулярными выражениями
```

```python
# загружаем в переменную df данные из исходного файла test_data.csv (создаем pandas dataframe из данных файла test_data.csv)
df = pd.read_csv('/content/test_data.csv')
df
```

```python
# в соответствии с заданием дополним таблицу следующими столбцами:
# greeting - напротив реплики с приветствием, для меток реплик с приветствием (принимает значение True, если обнаружено приветствие, иначе пустота)
# farewell - напротив реплики с прощанием, для меток реплик с прощанием (принимает значение True, если обнаружено приветствие, иначе пустота)
# name - имя менеджера, напротив реплик, где менеджер представил себя (принимает значение имени менеджера, если оно обнаружено, иначе путота)
# company - название компании, напротив реплик, где менеджер представил компанию (принимает значение названия компании, если оно обнаружено, иначе путота)
# kpi - показатель менеджера для каждого диалога (принимает значение True, если менеджер поздоровался и попрощался, иначе False)
df['greeting'] = ''
df['farewell'] = ''
df['name'] = ''
df['company'] = ''
df['kpi'] = ''
df
```

```python
# создаем паттерны для обнаружения приветствия менеджера в его репликах
# паттерны включают группу синонимов, которые можно как сокращать, так и расширять, в зависимости от поставленных требований

greeting_pattern =  r'\bдобрый день\b|\bдобрый\b|\bдень добрый\b|\bприветствую вас\b|\bдоброе утро\b|\
                    \bдоброго утра\b|\bутро доброе\b|\bдоброе\b|\bутра доброго\b|\bдоброго\b|\
                    \bдобрый вечер\b|\bвечер добрый\b|\bмое почтение\b|\bздравия желаю\b|\
                    \bпочет и уважение\b|\bрад слышать\b|\bрад вас слышать\b|\bрад тебя слышать\b|\
                    \bнизкий поклон\b|\bпривет\b|\bприветик\b|\bздравствуй\b|\bздравствуйте\b|\
                    \bздорово\b|\bздорова\b|\bздарово\b|\bздарова\b|\bдоброго здоровья\b|\
                    \bприветствую\b|\bассаламу алейкум\b|\bва алейкум салам\b|\bва алейкум ассалям\b|\
                    \bссаляму алейкум\b|\bсалам\b|\bсалют\b|\bсалютик\b|\bздрасте\b|\bздрасьте\b|\
                    \bздрасти\b|\bздрасьти\b|\bстрасте\b|\bсалам орлы\b|\bшалом\b|\bшелом\b|\
                    \bс добрым утром\b|\bдоброго времени суток\b|\bдоброго времени\b|\bсаламчики\b|\
                    \bсаламчик\b|\bхелло\b|\bхелоу\b|\bхеллоу\b|\bхай\b|\bхаюшки\b|\bвелкам\b|\
                    \bйоу\b|\bбонжур\b|\bприветец\b|\bхола\b|\bхаляйки\b|\bпревед\b|\bмабухай\b|\
                    \bку\b|\bалоха\b'
```

```python
# создаем паттерны для обнаружения прощания менеджера в его репликах
# паттерны включают группу синонимов, которые можно как сокращать, так и расширять, в зависимости от поставленных требований

farewell_pattern = r'\bдо свидания\b|\bдо свиданья\b|\bдо встречи\b|\bдо связи\b|\bпока\b|\
                    \bбывай\b|\bбудь здоров\b|\bбудьте здоровы\b|\bдо вечера\b|\bдо завтра\b|\
                    \bвсех благ\b|\bхорошего дня\b|\bхорошего утра\b|\bхорошего вечера\b|\
                    \bхороших выходных\b|\bприятных выходных\b|\bуспехов\b|\bвсего хорошего\b|\
                    \bудачи\b|\bвсего доброго\b|\bдо новых встреч\b|\bдо новой связи\b|\
                    \bвсего наилучшего\b|\bвсего лучшего\b|\bдо скорого\b|\bдо скорой встречи\b|\
                    \bдо скорой связи\b|\bсчастливо\b|\bувидимся\b|\bчесть имею\b|\
                    \bразрешите откланяться\b|\bпозвольте откланяться\b|\bразреши откланяться\b|\
                    \bпозволь откланяться\b|\bпозвольте попрощаться\b|\bпозволь попрощаться\b|\
                    \bразрешите попрощаться\b|\bразреши попрощаться\b|\bдо скорого свидания\b|\
                    \bдо скорого свиданья\b|\bсозвонимся\b|\bпересечемся\b|\bрад буду снова услышать\b|\
                    \bбуду рад снова услышать\b|\bбай\b|\bбай-бай\b|\bбай бай\b|\bчао\b|\
                    \bчесть имею кланяться\b|\bдо созвона\b|\bгуд бай\b|\bгуд-бай\b|\
                    \bауфидерзейн\b|\bариведерче\b|\bсайонара\b|\bата\b|\bадью\b|\bпокеда\b|\
                    \bпакеда\b|\bпакедова\b|\bпокедова\b|\bпокедово\b|\bуслышимся\b|\bреспект\b'
```

```python
# для обнаружения реплик с приветствием, именем менеджера и названием компании будем парсить только первые несколько реплик каждого диалога,
# во-первых потому-что данные реплики обязательно должны быть в самом начале диалога, во-вторых с целью сократить время и вычислительные мощности,
# аналогично для парсинга реплик с прощанием - поиск в последних двух репликах каждого диалога

# str_lower_begin - срез исходного фрейма по уловию "только первые реплики менеджера" для каждого диалога, весь текст приводим к нижнему регистру
str_lower_begin = df[(df.role == 'manager') & (df.line_n >= 0) & (df.line_n < 4)].text.str.lower()
# str_lower_end - срез исходного фрейма по уловию "только последние реплики менеджера" для каждого диалога, весь текст приводим к нижнему регистру
str_lower_end = df[(df.role == 'manager')].groupby('dlg_id', as_index = False)['text'].nth(n = [-2,-1]).str.lower()
```

```python
# создаем булеву маску greet для всех реплик менеджера в соответствии с паттернами greeting_pattern (парсинг приветствия)
greet_mask = str_lower_begin.str.contains(greeting_pattern)
# получаем список индексов исходного фрейма, где обнаружено приветствие
index_greet_true = str_lower_begin[greet_mask].index
# в колонку greeting напротив обнаруженных реплик с приветствием записываем значение 'True'
df.loc[index_greet_true, 'greeting'] = 'True'

# аналогично с репликами прощаний
farewell_mask = str_lower_end.str.contains(farewell_pattern)
index_farewell_true = str_lower_end[farewell_mask].index
df.loc[index_farewell_true, 'farewell'] = 'True'

# демонстрация того, что получилось на данном шаге, для среза только с репликами приветствия и прощания
df[(df.greeting == 'True') | (df.farewell == 'True')]
```

```python
# теперь заполняем колонку kpi (для тех диалогов, где было и приветствие и прощание указываем значение True, 
# в остальных случаях - False, значения заносятся напротив первой реплики менеджера, в каждом диалоге только одно значение)

# после того, как колонки greeting и farewell заполнены группируем фрейм по диалогам (индексы фрейма - это номера диалогов),
# агреггирем так, что бы для каждой группы вывести значения greeting и farewell (ниже результат продемонстрирован выводом на экран kpi_check)
kpi_check = df[df.role == 'manager'][['dlg_id', 'greeting', 'farewell']].groupby(['dlg_id'], as_index = False)[['greeting', 'farewell']].max()
kpi_check
```

```python
# создаем и выводим на экран маску для значений колонки kpi
kpi_mask = ((kpi_check.greeting == 'True') & (kpi_check.farewell == 'True'))
kpi_mask
```

```python
# в переменную kpi_index сохраняем индексы исходного фрейма, которые соответствуют всем первым репликам менеджера в каждом диалоге
# по этим индексам будем записывать значения в колонку kpi
kpi_index = df[df.role == 'manager'].groupby('dlg_id', as_index = False).nth(n = [0]).index
kpi_index
```

```python
# заполняем колонку kpi и выводим результат для демонстрации
df.loc[kpi_index, 'kpi'] = kpi_mask.to_list()
df[df.kpi != '']
```

```python
# приступаем к части парсинга имени менеджера и названия компании
# создаем объект morph как экземпляр класса pymorphy2.MorphAnalyzer() для морфологического анализа слов
morph = pymorphy2.MorphAnalyzer()
```

```python
# создаем список "стоп-слов", которые могут быть в репликах менеджера с представлением компании, данные слова будут учавствовать в формировании паттернов для регулярных выражений
# список включают группу синонимов, которые можно как сокращать, так и расширять, в зависимости от поставленных требований

stop_key = ['компания', 'компанию', 'компании', 'компаний', 'магазина', 'магазин',
            'магазинов', 'организация', 'организацию', 'организации', 'организаций', 
            'общество', 'предприятия', 'предприятие', 'предприятий', 'фирма',
            'фирму', 'фирмы', 'фирм', 'холдинг', 'агентство', 'агентства','агентств']
```

```python
# создаем паттерны для фильтрации реплик менеджера, в которых может содержаться название компании
# паттерны включают группу синонимов, которые можно как сокращать, так и расширять, в зависимости от поставленных требований

patterns_filters = [r'\bможно вас\b', r'\bя звоню\b', r'\bмы звоним\b', r'\bзвоню\b',
                    r'\bзвоним\b', r'\bудобно сейчас\b', r'\bудобно говорить\b',
                    r'\bвам сейчас удобно\b', r'\bсейчас удобно говорить\b',
                    r'\bвам удобно говорить\b', r'\bвы можете\b', r'\bвам удобно сейчас\b',
                    r'\bвы сейчас можете\b', r'\bвы можете говорить\b', r'\bпожалуйста\b',
                    r'\bне могли бы вы\b', r'\bмогли бы вы\b', r'\bкогда вам\b',
                    r'\bкогда удобно\b', r'\bмогу я\b', r'\bя могу\b',
                    r'\bменя зовут\b', r'\bслушаю вас\b', r'\bзовут меня\b', r'\bмое имя\b',
                    r'\bмы с вами\b', r'\bя с вами\b', r'\bс вами говорит\b',
                    r'\bмоя фамилия\b', r'\bзайму у вас\b', r'\bдобрый день\b',
                    r'\bдень добрый\b', r'\bприветствую вас\b', r'\bдоброе утро\b',
                    r'\bдоброго утра\b', r'\bутро доброе\b', r'\bутра доброго\b',
                    r'\bдобрый вечер\b', r'\bвечер добрый\b', r'\bмое почтение\b',
                    r'\bрад слышать\b', r'\bрад вас слышать\b', r'\bрад тебя слышать\b',
                    r'\bздравствуйте\b', r'\bприветствую\b', r'\bздрасте\b',
                    r'\bздрасьте\b', r'\bздрасти\b', r'\bздрасьти\b', r'\bэто\b',
                    r'\bс добрым утром\b', r'\bдоброго времени суток\b', r'\bя ваш',
                    r'\bваш', r'\bя прошу', r'\bпрошу', r'\bя вас', r'\bя хочу', r'\bхочу']
```

```python
# определяем функцию для парсинга имени компании, на вход функции подается реплика менеджера (text), которую необходимо проверить,
# возможную длину названия компании ограничим тремя элементами
def get_company_name(text):
  all_result = [] # здесь будут храниться имена компаний
  result = 'no name' # исходное значение для названия компании
  for f in stop_key: # итерация по списку стоп-слов, каждое из которых подставляется в паттерны для парсинга названия компании
    if re.findall(r'\b'+f+'\\b$', text) == [] and re.findall(r'\b'+f+'\\b', text) != []: # если паттерн с стоп-словом найден и этим стоп-слово не заканчивается реплика, то продолжаем

      if re.findall(r'(\b'+f+'\\b\s+\S+\s+да?и?\s+\S+\s+)(.+)', text) != []: # если найден паттерн после стоп-слова содержащий два элемента разделенные частицами "да" или "и", то это название компании
        result = re.findall(r'(\b'+f+'\\b\s+\S+\s+да?и?\s+\S+\s+)(.+)', text)[0][0].strip()
      else:
        first_parse = re.findall(r'(\b'+f+'\\b\s+\S+\s+)(.+)', text) # парсим первый элемент(слово или что-то другое) следующий сразу после стоп-слова, и парсим все, что следует за этим первым элементом
        second_parse = first_parse[0][1] # сохраняем строку, которая следует после первого элемента для ее дальнейшей фильтрации, анализа
        # first_parse[0][0] - первый элемент уже будет первой частью названия компании (или единственная его часть)
        #print(first_parse)
        #print(second_parse)

        for i in patterns_filters: # итерация по списку паттернов-фильтров
          pre_result = re.search(i, second_parse) # сохраненную ранее строку second_parse фильтруем от элементов, которые не могут быть частью названия компании
          if pre_result != None:
            #print(pre_result)
            start_del_index = pre_result.span()[0]
            #print(start_del_index)
            second_parse = second_parse[:start_del_index]
            #print(second_parse)

        second_parse_list = second_parse.split() # из отфильтрованной в предыдущем блоке строки формируем список элементов, разделенных пробелами
        if len(second_parse_list) != 0:
          check_name = second_parse_list[0]
          p = morph.parse(check_name)[0]
          # первый элемент этого списка сначало проверяем на морфологию, 
          # если это имя собственное в именителььном падеже, то это слово и все, 
          # что правее отскаем и берем за результат только первое слово, следующее за стоп-словом
          if ('Name' in p.tag) and ('nomn' in p.tag): 
            result = first_parse[0][0].strip()
          else:
            if len(second_parse_list) >= 2: # если в списке больше одного элемента, то проверяем морфологию двух перых элементов, остальные элементы не берем для анализа
              # грамеммы для второго слова из нашего списка сохраняем в переменную "p",
              # далее проверка нескольких возможных комбинаций двух слов из нашего списка для отсечения тех, которые не могут быть частью названия компании,
              # например, "добрый день, компания ЧИП менеджер Иван", или, "здравствуйте, магазин ДЕСТКИЙ МИР Иван", аналогично предыдущему блоку морфология по имени и падежу
              p = morph.parse(second_parse_list[1])[0]
              if ((second_parse_list[0] == 'менеджер' or second_parse_list[0] == 'я') and\
                  ('Name' in p.tag) and ('nomn' in p.tag)) or (second_parse_list[0] == 'я' and\
                                                               second_parse_list[1] == 'менеджер'):
                result = first_parse[0][0].strip() # если соответствия не нужным комбинациям были обнаружены, то в результат уходит только первый элемент после стоп-слова
              elif ('Name' in p.tag) and ('nomn' in p.tag): # проверяем второе слово из нашего списка на имя собственное и иминительный падеж (считается, что такая форма на третье позиции не может входить в название компании)
                result = first_parse[0][0]+second_parse_list[0] # тогда берем в результат только первый элемент после стоп-слова и первый элемент из нашего списка
              else:
                result = first_parse[0][0]+second_parse_list[0]+' '+second_parse_list[1] # иначе в результат берем все три элемента, это максимальная заданная нами длина названия компании
            else:
              result = first_parse[0][0]+second_parse_list[0] # если в нашем списке только один элемент, то конкатерируем его с первым элементов после стоп-слова, это и будет результатом
        else:
          result = first_parse[0][0].strip() # если наш список пуст, то в качестве названия компании берем только первый элемент после стоп-слова
    all_result.append(result) # аппендим все возможные названия компаний (потому что малая доля вероятности обнаружения нескольких паттернов в одной реплике всеже остается)
  return set(all_result) # результат обертываем множеством, чтобы избавиться от дублей
```

```python
# вызовем функцию для первых реплик менеджера из каждого диалога, для промежуточной демонстрации работы функции
company_name = str_lower_begin.apply(get_company_name)
company_name
```

```python
# создаем паттерны для обнаружения имени менеджера в его репликах
# паттерны включают группу синонимов, которые можно как сокращать, так и расширять, в зависимости от поставленных требований

patterns_get_name = [r'\bменя зовут\s+?\b\w{3,}\b\s*?(?:\s+\w{3,}\s*?)?(?:\s+\w{3,}\s?)?',
                     r'\bменя\s+?\b\w{3,}\b\s*?(?:\s+\w{3,}\s*?)?(?:\s+\w{3,}\s?)?',
                     r'\bзовут\s+?\b\w{3,}\b\s*?(?:\s+\w{3,}\s*?)?(?:\s+\w{3,}\s?)?',
                     r'\bя\s+?\b\w{3,}\b\s*?(?:\s+\w{3,}\s*?)?(?:\s+\w{3,}\s?)?',
                     r'\bэто\s+?\b\w{3,}\b\s*?(?:\s+\w{3,}\s*?)?(?:\s+\w{3,}\s?)?',
                     r'\bмое имя\s+?\b\w{3,}\b\s*?(?:\s+\w{3,}\s*?)?(?:\s+\w{3,}\s?)?',
                     r'\bмоя фамилия\s+?\b\w{3,}\b\s*?(?:\s+\w{3,}\s*?)?(?:\s+\w{3,}\s?)?',
                     r'\bс вами говорит\s+?\b\w{3,}\b\s*?(?:\s+\w{3,}\s*?)?(?:\s+\w{3,}\s?)?',
                     r'\bкомпания\s+?\b\w{3,}\b\s*? (?:и \b\w{3,}\b)?((?:\s*?\w{1,}\s*?)?(?:\s+\w{1,}\s*?)?(?:\s+\w{1,}\s?)?)',
                     r'\bменеджер\s+?\b\w{3,}\b\s*?(?:\s+\w{3,}\s*?)?(?:\s+\w{3,}\s?)?',
                     r'\bмагазин\s+?\b\w{3,}\b\s*? (?:и \b\w{3,}\b)?((?:\s*?\w{1,}\s*?)?(?:\s+\w{1,}\s*?)?(?:\s+\w{1,}\s?)?)',
                     r'\bмагазина\s+?\b\w{3,}\b\s*? (?:и \b\w{3,}\b)?((?:\s*?\w{1,}\s*?)?(?:\s+\w{1,}\s*?)?(?:\s+\w{1,}\s?)?)',
                     r'\bкомпании\s+?\b\w{3,}\b\s*? (?:и \b\w{3,}\b)?((?:\s*?\w{1,}\s*?)?(?:\s+\w{1,}\s*?)?(?:\s+\w{1,}\s?)?)']
```

```python
# определяем функцию для парсинга имени менеджера, на вход функции подаем реплику менеджера
def get_name(k):
  name = 'no name' # значение имени по умолчанию 
  for m in patterns_get_name: # итерация по списку паттернов
    result = re.findall(m, k) # поиск вхождений паттернов в реплику
    if result != []: # если вхождение есть, то создаем пустой список для добавления имени
      name = []
      for i in result: # итерация по списку обнаруженных вхождений
        n = i.split()  # каждый элемент списка (элемент списка это строка) переводим в список по признаку пробельного символа
        for h in n: # анализируем каждый элемент полученного списка по морфологии, если имя собственное, фамилия и(или) отчество, то это результат
          p = morph.parse(h)[0]
          if any(['Name' in p.tag, 'Surn' in p.tag, 'Patr' in p.tag]):
            name.append(p.word)
        if len(name) >= 1: # если хотябы один элемент имени обнаружен, то прекращаем итерацию обнаруженных вхождений
          break
      if len(name) >= 1: # если хотябы один элемент имени обнаружен, то прекращаем итерацию по списку паттернов
        break
  return name
```

```python
# вызовем функцию для первых реплик менеджера из каждого диалога, для промежуточной демонстрации работы функции
manager_name = str_lower_begin.apply(get_name)
manager_name
```

```python
# заполним колонку name и продемонстрируем результат, 
# заполнять будем позиции только напротив тех реплик, 
# где обнаружено имя менеджера, остальные позиции оставим пустыми
manager_mask = manager_name != 'no name'
final_manager_name = manager_name[manager_mask]
df.loc[final_manager_name.index, 'name'] = final_manager_name
df[df.name != ''] # демонстрация результатов по срезу только для диалогов, где обнаружены имена менеджеров
```
